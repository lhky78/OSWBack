```python
url='https://search.shopping.naver.com/search/all.nhn?query=%EB%85%B8%ED%8A%B8%EB%B6%81&frm=NVSCPRO'
```

```python
driver=webdriver.Chrome(r'C:\Users\lhky0\Desktop\db\chromedriver_win32\chromedriver')
driver.implicitly_wait(3)
#selenium의 모든 자원이 로드될 때까지 기다림
driver.get(url)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

table = soup.find(id='content')
name = table.find_all("div",{"class":"tit"})
price = table.find_all("span",{"class":"num _price_reload"})
review = table.find_all("span",{"class":"num _price_reload"})
spec = table.find_all("span",{"class":"detail"})
```

```python
results=[[0 for x in range(3)] for y in range(1080)] # name, price, spec
```

```python
idx=0
sum=0
for i in range(10000000) :
    
    results[i][0]=name[idx].get_text()
    results[i][1]=price[idx].get_text()
    results[i][2]=spec[idx].get_text()
    print(idx, name[idx].get_text(), price[idx].get_text())
    print(spec[idx].get_text())
    #print(vals[idx].get_text().strip(),
     #     scores[idx].a.find_all("",{"class":""})[0].get_text())
    idx+=1
    print("")
   
    
    if(idx==len(spec)):
        idx=0
        driver.find_element_by_xpath(xpath).click()
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        table = soup.find(id="content")
        name = table.find_all("div",{"class":"tit"})
        price = table.find_all("span",{"class":"num _price_reload"})
        review = table.find_all("span",{"class":"num _price_reload"})
        spec = table.find_all("span",{"class":"detail"})
        sum+=len(spec)
        print(len(name), len(price), len(spec),sum)
        
        if(len(spec)==0):
            continue;

        print("")
        
        
```
